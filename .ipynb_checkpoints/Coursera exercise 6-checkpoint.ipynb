# Support Vector Machines

In this notebook three different datasets of progressive complexity are used to train various support vector machine models.

At the end, grid search is manually performed to find the optimal parameters for use in our model, and then this is compared to a gridsearch using the highly optimised Sklearn GridsearchCV model.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.io import loadmat

datafile = loadmat('machine-learning-ex6/ex6/ex6data1.mat', struct_as_record=False)

X = datafile['X']
y = datafile['y']

print("The shape of X is: {}".format(X.shape))
print("The shape of y is: {}".format(y.shape))
```

### Visualise dataset

```python
X = np.array(X)
y = np.array(y.flatten())
class_one = (y == 0)
class_two = (y == 1)
```

```python
plt.figure()
plt.scatter(X[class_one, 0], X[class_one, 1], c='r', marker='x')
plt.scatter(X[class_two, 0], X[class_two, 1], c='b', marker='o')
plt.show()
```

### Normalise features and visualise again

```python
def feature_normalise(X):
    X_norm = (X - (np.mean(X, axis=0))) / np.std(X, axis = 0)
    return X_norm
```

```python
X_norm = feature_normalise(X)
```

```python
plt.figure()
plt.scatter(X_norm[(y == 0), 0], X_norm[(y == 0), 1], c='r', marker='x')
plt.scatter(X_norm[(y == 1), 0], X_norm[(y == 1), 1], c='b', marker='o')
plt.show()
```

### Support Vector Machine Classification

```python
from sklearn.svm import SVC

models = (SVC(C=1.0, kernel='linear'), SVC(C=100.0, kernel='linear'))
models = (clf.fit(X_norm, y) for clf in models)

titles = ('SVC with linear kernel, C = 1.0',
          'SVC with linear kernel, C = 100.0')
```

The C value is a positive value that is a measure of how much we penalise misclassified training examples. If we have a large value of C, our SVM will work to classify every single training example correctly. If on the other hand we choose a low value of C, our SVM will form a classification boundary that works for the majority of the data. The result of this is that large outliers or singled out training examples will be ignored.

```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
X0, X1 = X_norm[:, 0], X_norm[:, 1]
xx, yy = make_meshgrid(X0, X1)

fig, sub = plt.subplots(2,figsize=(8,12))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title(title)
plt.show()
```

Notice that when C = 1.0, the outlier to the far left is not included within the classification boundary. When C = 100.0, this point is then included within the classification boundary.

The value of C chosen is crucial for our SVM. On the extremeties of our choice for C:

#### Create a custom gaussian kernel for our Support Vector Machine - NOT WORKING YET - to be debugged...

The gaussian kernel function is defined as:

$ K_{Gaussian}(x^{(i)}, x^{(j)}) = exp \Big( - \displaystyle\frac{||x^{(i)} - x^{(j)} ||^{2}}{2\sigma^{2}} \Big) $

$ K_{Gaussian}(x^{(i)}, x^{(j)}) = exp \Big( - \displaystyle\frac{ \sum_{k=1}^{n} (x^{(i)} - x^{(j)})^{2}}{2\sigma^{2}} \Big) $

```python
def build_k_gaussian(sigma):

    def k_gaussian(_x1, _x2):
        normsq = np.square(np.linalg.norm(_x1 - _x2))
        return np.exp(- normsq / (2 * np.square(sigma)))

    return k_gaussian
```

```python
X_norm = feature_normalise(X)
clf = SVC(C=1.0, kernel=build_k_gaussian(sigma=2))
clf = clf.fit(X_norm, y)

X0, X1 = X_norm[:, 0], X_norm[:, 1]
xx, yy = make_meshgrid(X0, X1)

fig, ax = plt.subplots(figsize=(8,12))
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title(title)
plt.show()
```

### Dataset 3 - Optimising the hyperparameters of our model

```python
datafile3 = loadmat('machine-learning-ex6/ex6/ex6data3.mat', struct_as_record=False)

X = datafile3['X']
X_val = datafile3['Xval']
y = datafile3['y'].flatten()
y_val = datafile3['yval'].flatten()

print("The shape of X is: {}".format(X.shape))
print("The shape of X validation is: {}".format(X_val.shape))
print("The shape of y is: {}".format(y.shape))
print("The shape of y validation is: {}".format(y_val.shape))
```

```python
plt.figure()
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='r', marker='x')
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='b', marker='o')
plt.show()
```

#### Perform a manual grid-search for the optimal values of C and gamma.

```python
C_values = [0.01, 0.03, 0.1, 0.3, 1.0, 3.0, 10.0, 30.0]
gamma_values = [0.01, 0.03, 0.1, 0.3, 1.0, 3.0, 10.0, 30.0]
parameter_grid = [(a, b) for a in C_values for b in gamma_values]
scores = []

for C, gamma in parameter_grid:
    clf = SVC(kernel='rbf', gamma=gamma, C=C).fit(X, y)
    score = clf.score(X_val, y_val)
    scores.append([score, C, gamma])
scores = np.array(scores)
```

```python
top_model = scores[scores[:, 0].argmax()]
print("The optimal model has: C = {0} and gamma = {1}, which scored {2}".format(top_model[1], 
                                                                                top_model[2],
                                                                                top_model[0]))
```

##### Visualise our model with the training data

```python
best_clf = SVC(kernel='rbf', gamma=30.0, C=3.0).fit(X, y)
```

```python
X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

fig, ax = plt.subplots(figsize=(8,10))
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plot_contours(ax, best_clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.5)
ax.scatter(X0[y == 0], X1[y == 0], c='b', cmap=plt.cm.coolwarm, s=25, edgecolors='k', marker='o')
ax.scatter(X0[y == 1], X1[y == 1], c='brown', cmap=plt.cm.coolwarm, s=25, edgecolors='k', marker='x')
ax.set_xlim(X0.min(), X0.max())
ax.set_ylim(X1.min(), X1.max())
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title("Optimal Gaussian SVM: C = 3.0, gamma = 30.0")
plt.show()
```

This was reasonably successful, however it can be done much easier using sklearn gridsearch...

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Create my estimator and prepare the parameter grid dictionary
params_dict = {"C": np.logspace(-1, 3, 100), "gamma": np.linspace(0.0001, 10, 100)}
svm_clf = SVC(kernel="rbf")

# Fit the grid search
grid_search = GridSearchCV(estimator=svm_clf, param_grid=params_dict)
grid_search.fit(X, y)

# Capture and fit the best estimator from across the grid search# Capture 
best_svm = grid_search.best_estimator_
best_svm.fit(X, y)
```

```python
print("This grid search optimised model has a validation score of: {} ".format(best_svm.score(X_val, y_val)))
```

```python
# Generate a set of points that form a grid over feature space# Generat 
x1s = np.linspace(min(X[:, 0]), max(X[:, 0]), 600)
x2s = np.linspace(min(X[:, 1]), max(X[:, 1]), 600)
mesh_grid = np.array([[x1, x2] for x1 in x1s for x2 in x2s])

# Compute decision function for each point, keep those which are close to the boundary
dists = best_svm.decision_function(mesh_grid)
bounds = np.array([pt for pt, dist in zip(mesh_grid, dists) if abs(dist)<0.05])

# Visualize the decision boundary
fig, ax = plt.subplots()
ax.scatter(X[y == 0, 0], X[y == 0, 1], color='r', marker='x')
ax.scatter(X[y == 1, 0], X[y == 1, 1], color='b', marker='o')
ax.scatter(bounds[:, 0], bounds[:, 1], color="g", s=3)
ax.set_title("Grid-Search Optimized SVM")
plt.show()
```

Interestingly, our manually optimised grid search actually scored higher on the validation set, with a score of 0.965, as opposed to the sklearn gridsearch model, which scored 0.96.

We'll put this down to luck... Usually the convenience and ease of use of the sklearn grid search, with adequately chosen values, will yield a better optimisation in a faster timescale than one searched manually.
