### Produce feedforward function for obtaining y predictions

```python
def predict_y(X, Theta_1, Theta_2):
    # add a column of 1's to the training data for the bias unit
    bias = np.ones((X.shape[0], 1))
    a_1 = np.column_stack((bias, X))
    z_2 = np.dot(Theta_1, a_1.T)
    a_2 = (1.0 / (1 + np.exp(-z_2)))
        
    # add a bias row to the second layer data
    bias = np.ones(a_2.shape[1])
    a_2 = np.row_stack((bias, a_2))
        
    z_3 = np.dot(Theta_2, a_2)
    a_3 = (1.0 / (1 + np.exp(-z_3)))
    a_3 = a_3.T
        
    # return the output labels in an array of shape (predict_no, 10)
    return a_3
```

Custom function for one-hot encoding our y labels. It should be noted that this function caused huge difficulty in the beginning and should not be used in favour of the pandas built-in function for doing this - I simply created this for my own education in one-hot encoding using only numpy.

```python
def one_hot_encode(one_dim_array):
    """ Convert our integer outputs into vectors with 1 or 0, where 1 is output 
        at the index relating to the specified output class """
    a = one_dim_array.flatten()
    b = np.zeros((a.shape[0], np.unique(y).shape[0] + 1))
    b[np.arange(a.shape[0]), a] = 1
    return b

# Huge point to note for future - this can be done very easily using pandas:
# y_enc = pd.get_dummies(y.ravel()).as_matrix()
```

```python
y_encoded = one_hot_encode(y)
print("Initial one-hot encoded y labels shape: {}".format(y_encoded.shape))
print(y_encoded[0])

# remove column 1 from y as it is not needed (octave does not have 0 indexing)
y_encoded = y_encoded[:, 1:]
print("\nFinalised one-hot encoded y labels shape: {}".format(y_encoded.shape))
print(y_encoded[0])
```

### Multilayer Neural Network Cost Functions

### Cost Function 
$$ J(\theta) = \frac{1}{m}\sum_{i=1}^{m}\sum_{k=1}^{K}\big[-y^{(i)}_{k}\, log\,(( h_\theta\,(x^{(i)}))_k)-(1-y^{(i)}_k)\,log\,(1-h_\theta(x^{(i)}))_k)\big]$$
 
### Regularised Cost Function
$$ J(\theta) = \frac{1}{m}\sum_{i=1}^{m}\sum_{k=1}^{K}\bigg[-y^{(i)}_{k}\, log\,(( h_\theta\,(x^{(i)}))_k)-(1-y^{(i)}_k)\,log\,(1-h_\theta(x^{(i)}))_k)\bigg] + \frac{\lambda}{2m}\bigg[\sum_{j=1}^{25}\sum_{k=1}^{400}(\Theta_{j,k}^{(1)})^2+\sum_{j=1}^{10}\sum_{k=1}^{25}(\Theta_{j,k}^{(2)})^2\bigg]$$

We can calculate the cost one of many ways - either using 'for loops' or more efficient vectorisation. With the former, we iteratively go through our samples and add up the costs, as shown in the cost function defined below.

```python
def cost_calculation(X, y, Theta_1, Theta_2, epsilon = 1e-12):
    m = X.shape[0]
    y_pred = predict_y(X, Theta_1, Theta_2)
    cost = 0
    for i in range(m):
        cost_sum = ((-y[i].dot(np.log(y_pred[i]+epsilon))) - ((1.0 - y[i]).dot(np.log((1.0 - y_pred[i])+epsilon))))
        cost += cost_sum
    final_cost = cost / m
    return final_cost
```

```python
cost = cost_calculation(X, y_encoded, Theta1, Theta2)
print("The calculated cost using the provided pre-optimised weights: {}\n".format(cost))
print("Expected cost from Andrew Ng's Stanford course is about 0.287.")
```

A better option for calculating the cost, as opposed to for loops, is through vectorised operations in numpy. This can be more confusing for those unfamiliar with Advanced Linear Algebra and numpy operations, however it is much faster and more efficient, so is worth the extra work in understanding it. An implementation using this technique is given below.

```python
def vectorised_cost_calculation(X, y, Theta_1, Theta_2, epsilon = 1e-12):
    m = float(X.shape[0])
    y_pred = predict_y(X, Theta_1, Theta_2)
    cost_matrix = ((-y.dot(np.log((y_pred + epsilon).T))) - ((1.0 - y).dot(np.log(((1.0 - y_pred) + epsilon).T)))) / m
    # this matrix has unwanted terms - only the diagonal elements are wanted for our sum
    # these can be summed as required using np.trace() on an np formatted array
    cost = np.trace(cost_matrix)
    
    # alternative easy way of writing the cost - using matrix multiplication rather than np.dot
    #cost = -1*(1/m)*np.sum(np.log(y_pred)*(y) + np.log(1-y_pred)*(1 - y)) 
    return cost
```

```python
alt_cost = vectorised_cost_calculation(X, y_encoded, Theta1, Theta2)
print("Calculated cost using vectorised cost function: {}\n".format(alt_cost))
```

We obtain precisely the same value for the cost, albeit done more quickly, efficiently and with less code!

```python
def regularised_cost(X, y, Theta_1, Theta_2, reg_term, epsilon = 1e-12):
    m = float(X.shape[0])
    y_pred = predict_y(X, Theta_1, Theta_2)
    cost_matrix = (-y * np.log((y_pred + epsilon))) - ((1.0 - y)*np.log(((1.0 - y_pred) + epsilon)))
    cost = np.sum(cost_matrix) / m
    
    # regularisation does not include the first bias term of weights matrices
    reg_sum = (reg_term / (2*m)) * (np.sum(Theta_1[:, 1:] ** 2) + np.sum(Theta_2[:, 1:] ** 2))
    cost += reg_sum
    return cost
```

```python
final_cost = regularised_cost(X, y_encoded, Theta1, Theta2, reg_term=1.0)
print("Calculated regularised cost: {}\n".format(final_cost))
print("Expected cost from Andrew Ng's course exercises is 0.383.")
```

### Multilayer Neural Network Backpropagation

#### Sigmoid function
$$ g(z) = \frac{1}{1+e^{-z}}$$

#### Sigmoid gradient
$$ g'(z) = g(z)(1 - g(z))$$

#### Terms for backpropagation

$$ δ^{(l)} = ((Θ^{(l)})^T δ^{(l+1)}) .∗ a^{(l)} .∗ (1−a^{(l)}) $$

Where L is our number of layers and $ a^{(L)} $ is the vector of outputs of our activation units for the last layer. Also remember that the notation $ .* $ means element-wise multiplication of matrices. 

For the first term $ δ^{(L)} $ we simply calculate: $ δ^{(L)} = a^{(L)} - y^{(i)} $.

There is also no difference term for the input layer, $ δ^{(1)} $, since this layer simply receives the input features and thus there are no errors associated here.

These delta terms are actually the partial derivatives of the cost function J.

#### Regularisation for backpropagation

Where L is our number of layers and $ a^{(L)} $ is the vector of outputs of our activation units for the last layer. Also remember that the notation $ .* $ means element-wise multiplication of matrices. 

For the first term $ δ^{(L)} $ we simply calculate: $ δ^{(L)} = a^{(L)} - y^{(i)} $.

For $ j \neq 0 $:

$$ D_{i,j}^{(l)} = \frac{1}{m}(\Delta_{i,j}^{(l)} + \lambda\Theta_{i,j}^{(l)}) $$

and for $ j = 0 $:

$$ D_{i,j}^{(l)} = \frac{1}{m}(\Delta_{i,j}^{(l)}) $$

From these equations above, its clear that we do not use the first column of the weights matrix for regularisation, since this is the bias term. We can then subtract these associated gradient values from our weight matrices in order to optimise our cost function through backpropagation.

```python
def backpropagation(X, y, Theta_1, Theta_2, reg_term):
    # add a column of 1's to the training data for the bias unit
    bias_col = np.ones((X.shape[0], 1))
    a_1 = np.column_stack((bias_col, X))
    z_2 = np.dot(Theta_1, a_1.T)
    a_2 = (1.0 / (1 + np.exp(-z_2)))
    
    # add a bias row to the second layer data
    bias_row = np.ones(a_2.shape[1])
    a_2 = np.row_stack((bias_row, a_2))
        
    z_3 = np.dot(Theta_2, a_2)
    a_3 = (1.0 / (1 + np.exp(-z_3)))
    a_3 = a_3.T
    
    # step back from right to left, obtaining d values
    # d3 equal to a_3 predictions and actual y values
    d_3 = a_3 - y
    
    # we dont want first column of theta_2 - this is from bias term
    d_2 = (np.dot(Theta_2[:, 1:].T, d_3.T)) * sigmoid_gradient(z_2)
    
    # we have no d1 term - all layer 1 receives is our input features so there is no
    # error associated with it
    
    m = y.shape[0]
    
    # calculations for gradients using d terms
    grad_2 = np.dot(d_3.T, a_2.T)
    grad_1 = np.dot(d_2, a_1)
    
    # set first columns of Theta1 and Theta2 to zero for regularisation
    Theta_1[:, 0] = 0
    Theta_2[:, 0] = 0 
    
    # calculate final delta values
    delta_2 = (grad_2 + (reg_term*Theta_2)) / m
    
    delta_1 = (grad_1 + (reg_term*Theta_1)) / m
        
    return delta_1, delta_2

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_gradient(z):
    return sigmoid(z)*(1 - sigmoid(z))
```

```python
D_1, D_2 = backpropagation(X, y_encoded, Theta1, Theta2, reg_term = 1.0)
print(D_1.shape)
print(D_2.shape)
```

```python
class Simple_Neural_Network(object):
    """ A simple neural network with an input layer (400 units), hidden layer (25 units) and 
        output layer (10 units). Sigmoid is used as the activation function.
        Returns an array of size 10, with the class of digit from 0 - 9
        indicated with a 1 in the corresponding index of that array.
    """

    def __init__(self, input_units = 400, hidden_units = 25, output_units = 10, learn_rate = 0.01, 
                 reg_term = 0, iterations = 1000, random_seed = 1):
        self.input_units = input_units
        self.hidden_units = hidden_units
        self.output_units = output_units
        self.learn_rate = learn_rate
        self.reg_term = reg_term
        self.iterations = iterations
        
        # initialise randomised normal distributed weights for our network
        random_gen = np.random.RandomState(random_seed)
        self.theta1 = random_gen.normal(loc = 0.0, scale = 0.01, size =(hidden_units, input_units + 1))
        self.theta2 = random_gen.normal(loc = 0.0, scale = 0.01, size =(output_units, hidden_units + 1))

        
    def fit(self, X, y, check_gradients=False):
        J_history = []
        for i in range(self.iterations):
            a_1, a_2, z_2, a_3 = self._feedforward(X)
            D_1, D_2 = self._backpropagation(X, y, a_1, a_2, z_2, a_3) 
            cost = self._calculate_cost(X, y, a_3)
            J_history.append(cost)
            
            if check_gradients:
                delta_vect = np.concatenate([D_1.flatten(), D_2.flatten()])
                self._gradient_check(X, y, a_3, delta_vect)
            
        return J_history
    
    
    def predict(self, X_pred):
        a_1, a_2, z_2, a_3 = self._feedforward(X)
        return a_3
       
        
    def _feedforward(self, X):
        # add a column of 1's to the training data for the bias unit
        bias = np.ones((X.shape[0], 1))
        a_1 = np.column_stack((bias, X))
        z_2 = np.dot(self.theta1, a_1.T)
        a_2 = self._sigmoid(z_2)
        
        # add a bias row to the second layer data
        bias = np.ones(a_2.shape[1])
        a_2 = np.row_stack((bias, a_2))
        
        z_3 = np.dot(self.theta2, a_2)
        a_3 = self._sigmoid(z_3)
        a_3 = a_3.T
        
        return a_1, a_2, z_2, a_3
    
    
    def _calculate_cost(self, X, y, y_pred, epsilon = 1e-12):
        m = float(X.shape[0])
        cost_matrix = (-y * np.log((y_pred + epsilon))) - ((1.0 - y)*np.log(((1.0 - y_pred) + epsilon)))
        cost = np.sum(cost_matrix) / m
    
        # regularisation does not include the first bias term of weights matrices
        reg_sum = (self.reg_term / (2*m)) * (np.sum(self.theta1[:, 1:] ** 2) + np.sum(self.theta2[:, 1:] ** 2))
        cost += reg_sum
        return cost

    
    def _backpropagation(self, X, y, a_1, a_2, z_2, a_3):
        # step back from right to left, obtaining d values
        # d3 equal to a_3 predictions and actual y values
        d_3 = a_3 - y
    
        # we dont want first column of theta_2 - this is from bias term
        d_2 = (np.dot(self.theta2[:, 1:].T, d_3.T)) * self._sigmoid_gradient(z_2)
    
        # we have no d1 term - all layer 1 receives is our input features so there is no
        # error associated with it
    
        m = y.shape[0]
    
        # calculations for gradients using d terms
        grad_2 = np.dot(d_3.T, a_2.T)
        grad_1 = np.dot(d_2, a_1)
    
        # set first columns of Theta1 and Theta2 to zero for regularisation
        reg_theta_1 = np.copy(self.theta1)
        reg_theta_1[:, 0] = 0
        
        reg_theta_2 = np.copy(self.theta2)
        reg_theta_2[:, 0] = 0 
    
        # calculate final delta values
        delta_2 = (grad_2 + (self.reg_term*reg_theta_2)) / m
        delta_1 = (grad_1 + (self.reg_term*reg_theta_1)) / m
        
        # update theta matrices using delta and learning rate
        self.theta1 -= self.learn_rate * delta_1
        self.theta2 -= self.learn_rate * delta_2
        
        return delta_1, delta_2
        
        
    def _sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))
    
    
    def _sigmoid_gradient(self, z):
        return sigmoid(z)*(1 - sigmoid(z))
    
    
    def _gradient_check(self, X, y, y_pred, delta_vect, epsilon=10e-4):
        """ Function to check the gradients calculated during backpropagation.
            WARNING - only run gradient check with a very low number of iterations
            of the neural network, since the calculations will take a long time """
        
        # flatten weight matrices and combine into one large theta vector
        theta_vect = np.concatenate([self.theta1.flatten(), self.theta2.flatten()]) 
        n = theta_vect.shape[0]
        m = y.shape[0]
        grad_approx = np.zeros(theta_vect.shape[0])
        
        for i in range(n):
            theta_plus = theta_vect[:]
            theta_plus[i] + epsilon
            theta_minus = theta_vect[:]
            theta_minus[i] - epsilon
            plus_cost = self._grad_check_cost(X, y, y_pred, theta_plus)
            minus_cost = self._grad_check_cost(X, y, y_pred, theta_plus)
            grad_approx[i] =  (plus_cost - minus_cost) / (2*epsilon)
            
        dif_vect = delta_vect - grad_approx
        print(dif_vect[:100])
            
            
    def _grad_check_cost(self, X, y, y_pred, theta_vect, epsilon=1e-12):
        """ Modified cost function to determine cost using a single flattened theta 
            vector, which is used during gradient checking. """
        
        # reshape our theta vect into seperate theta matrices
        theta_1 = theta_vect[:10025].reshape((25, 401))
        theta_2 = theta_vect[10025:].reshape((10, 26))
        m = float(X.shape[0])
        cost_matrix = (-y * np.log((y_pred + epsilon))) - ((1.0 - y)*np.log(((1.0 - y_pred) + epsilon)))
        cost = np.sum(cost_matrix) / m
        reg_sum = (self.reg_term / (2*m)) * (np.sum(self.theta1[:, 1:] ** 2) + np.sum(self.theta2[:, 1:] ** 2))
        cost += reg_sum
        return cost
```

```python
# checking of the gradients of our model
neural_network_model = Simple_Neural_Network(iterations=1)
neural_network_model.fit(X, y_encoded, check_gradients=True)
```

From the difference results shown above (approximate gradient - backpropagation calculated gradients), we can see that the differences are very minor (all agree to at least 4 s.f.). This gives us confidence that the backpropagation algorithm in our code is working correctly.
