### Regularised Linear Regression Cost Function

For regularised linear regression, our goal is to minimise a cost function through optimisation of our model parameters, $ \theta $. Regularisation is a term that we add to the original linear regression cost function that provides a degree of control over underfitting and overfitting (bias and variance). The regularisation term can be thought of as a penalty factor on the overall cost, $ J(\theta) $. The larger the magnitudes of the parameter vector terms in $ \theta $, the larger the penalty factor. 

The regularised cost function we'll be using is as follows:

$ J(\theta) = \displaystyle\frac{1}{2m} \Big( \displaystyle\sum_{i=1}^{m} \Big(h_{\theta}(x^{(i)}) - y^{(i)} \Big)^2 \Big) + \displaystyle\frac{\lambda}{2m}\Big(\displaystyle\sum_{j=1}^{n}\theta_{j}^{2} \Big)  $

An important factor is that we do not regularise the bias parameter term, $ \theta_{0} $.

```python
def net_input(X, theta):
    return np.dot(X, theta)

def add_bias(X):
    bias = np.ones((X.shape[0], 1))
    X_bias = np.column_stack((bias, X))
    return X_bias

def linear_regression(X, y, theta, reg_term=1.0, add_bias=False):
    if add_bias:
        X = add_bias(X)
    m = y.shape[0]
    hx = net_input(X, theta)
    errors = hx - y.reshape((y.shape[0],))
    
    # calculate reg cost
    basic_cost = ((errors**2).sum()) / (2.0*m)
    reg_term = (reg_term / (2.0*m)) * np.sum(theta[1:]**2)
    cost = basic_cost + reg_term
    
    # calculate reg gradients for optimisation
    grads = np.dot(errors, X) / m
    grad_reg_term = (reg_term/m)*theta[1:] 
    grads[:1] += grad_reg_term
    return cost, grads
```

```python
# add a column of 1's to the training data for the bias unit
bias = np.ones((X.shape[0], 1))
X_bias = np.column_stack((bias, X))
theta = np.ones(X_bias.shape[1])

cost, gradients = linear_regression(X_bias, y, theta, reg_term = 1.0, add_bias=False)
print("The cost returned was: {0}. Expected 303.993 from Stanford course exercise.".format(cost))
print("The gradients are: {0}. Expected [-15.30, 598.25] from course exercises.".format(gradients))
```

### Linear regression model

```python
class linear_regression(object):
    
    def __init__(self, learn_rate, iterations, reg_term):
        self.learn_rate = learn_rate
        self.iterations = iterations
        self.reg_term = reg_term
        
    def add_bias(self, X):
        bias = np.ones((X.shape[0], 1))
        X = np.column_stack((bias, X))
        return X
    
    def initialise_weights(self, X, ones=False):
        if ones:
            self.theta = np.ones(X.shape[1])
        else: 
            random_gen = np.random.RandomState(1)
            self.theta = random_gen.normal(loc = 0.0, scale = 0.01, size =(X.shape[1]))
        return self
            
    def net_input(self, X):
        return np.dot(X, self.theta)
       
    def fit(self, X, y, add_bias=False):
        if add_bias:
            X = self.add_bias(X)
        self.initialise_weights(X)
        m = y.shape[0]
        
        costs = []
        for cycle in range(self.iterations):
            hx = self.net_input(X)
            errors = hx - y.reshape((y.shape[0],))
    
            # calculate reg cost
            basic_cost = ((errors**2).sum()) / (2.0*m)
            reg_term = (self.reg_term / (2.0*m)) * np.sum(self.theta[1:]**2)
            cost = basic_cost + self.reg_term
            costs.append(cost)
    
            # calculate reg gradients for optimisation
            grads = X.T.dot(errors)
            grad_reg_term = (self.reg_term/m)*self.theta[1:] 
            grads[1:] += grad_reg_term
            self.theta -= self.learn_rate*grads
        return costs, self.theta
    
    def validation_error(self, X, y, add_bias=False):
        if add_bias:
            X = self.add_bias(X)
        m = y.shape[0]
        costs = []
        for cycle in range(self.iterations):
            hx = self.net_input(X)
            errors = hx - y.reshape((y.shape[0],))
            # calculate reg cost
            basic_cost = ((errors**2).sum()) / (2.0*m)
            reg_term = (self.reg_term / (2.0*m)) * np.sum(self.theta[1:]**2)
            cost = basic_cost + self.reg_term
            costs.append(cost)
        return costs
```

```python
regressor = linear_regression(learn_rate=0.00015, iterations=3000, reg_term=0.0)
costs, weights = regressor.fit(X, y, add_bias=True)
```

#### Visualisation of the cost over iterations of training

```python
plt.figure()
plt.plot(range(1, len(costs) + 1), costs)
plt.ylabel("Cost")
plt.xlabel("Training iterations (Epochs)")
plt.show()
```

### Generation of training curves - training set cost and validation set cost

Generation of training curves using different numbers of data samples for training and validation sets

```python
trg_entries = []
val_entries = []

for m in range(3, y.shape[0]):
    regressor = linear_regression(learn_rate=0.00015, iterations=2000, reg_term=0.0)
    trg_costs, weights = regressor.fit(X[:m], y[:m], add_bias=True)
    trg_entry = [m, trg_costs[-1]]
    trg_entries.append(trg_entry)
    #dict_key = "{0} samples".format(m)
    #learning_curve[dict_key] = costs[-1]
    
    val_costs = regressor.validation_error(X_val[:m], y_val[:m], add_bias=True)
    val_entry = [m, val_costs[-1]]
    val_entries.append(val_entry)
```

```python
trg_entries = np.array(trg_entries)
val_entries = np.array(val_entries)
```

```python
plt.figure()
plt.plot(trg_entries[:, 0], trg_entries[:, 1], marker='x')
plt.plot(trg_entries[:, 0], val_entries[:, 1], marker='o')
plt.xlabel("Number of training examples")
plt.ylabel("Error")
plt.legend(["Training", "Validation"])
plt.show()
```

Both training set and validation set are high, despite having a large number of iterations (2000). This is indicative of a high bias problem in our model. In other words, the linear regression line we are trying to use to approximate our data is not complex enough for the true relationship that exists. This can be solved by forming a more complex model, such as polynomial features.

--- 
### Improvement through addition of polynomial features

We'll add polynomial features to our model, and additionally, before applying these features we'll normalise the data with mean 0 and standard deviation of 1.

```python
def poly_features(X, polynomial_degree):
    out = np.copy(X)
    for i in range(2, polynomial_degree + 1):
        feature_column = X**i
        out = np.column_stack((out, feature_column))
    return out

def feature_normalise(X):
    X_norm = (X - (np.mean(X, axis=0))) / np.std(X, axis = 0)
    return X_norm
```

```python
X_norm, X_val_norm, X_test_norm = feature_normalise(X), feature_normalise(X_val), feature_normalise(X_test)
X_poly, X_val_poly, X_test_poly = poly_features(X_norm, 8), poly_features(X_val_norm, 8), poly_features(X_test_norm, 8)
print(X_poly.shape, X_val_poly.shape, X_test_poly.shape)
```

```python
regressor = linear_regression(learn_rate=0.00015, iterations=3000, reg_term=0.0)
costs_poly, weights_poly = regressor.fit(X_poly, y, add_bias=True)
```

```python
plt.figure()
plt.plot(range(1, len(costs_poly) + 1), costs_poly)
plt.title("Polynomial Regression")
plt.ylabel("Cost")
plt.xlabel("Training iterations (Epochs)")
plt.show()
```

```python
plt.figure()
plt.scatter(X_poly[:, 0], y, marker='x')
plt.xlabel("Change in water level (Normalised)")
plt.ylabel("Water flowing out of the dam")
# y = mx + c
plt.scatter(X_poly[:, 0], np.dot(X_poly, weights_poly[1:]) + weights_poly[0], c='r', alpha=0.5)
plt.legend(["Training data", "Polynomial fitment (n=8)"])
plt.show()
```

Clearly, with the addition of polynomial features, our model fits the training data much better.

```python
trg_entries_poly = []
val_entries_poly = []

for m in range(3, y.shape[0]):
    regressor = linear_regression(learn_rate=0.00015, iterations=2000, reg_term=0.0)
    trg_costs, weights = regressor.fit(X_poly[:m], y[:m], add_bias=True)
    trg_entry = [m, trg_costs[-1]]
    trg_entries_poly.append(trg_entry) 
    val_costs = regressor.validation_error(X_val_poly[:m], y_val[:m], add_bias=True)
    val_entry = [m, val_costs[-1]]
    val_entries_poly.append(val_entry)

trg_entries_poly = np.array(trg_entries_poly)
val_entries_poly = np.array(val_entries_poly)
```

```python
plt.figure()
plt.plot(trg_entries_poly[:, 0], trg_entries_poly[:, 1], marker='x')
plt.plot(val_entries_poly[:, 0], val_entries_poly[:, 1], marker='o')
plt.title("Learning curve for polynomial regression (n=8)")
plt.xlabel("Number of training examples")
plt.ylabel("Error")
plt.legend(["Training", "Validation"])
plt.show()
```

This is an improvement, however our validation error is still moderately high. This is indicative of high variance (overfitting), since our training error is low, whilst our validation error is high.

We can improve on this through regularisation.
