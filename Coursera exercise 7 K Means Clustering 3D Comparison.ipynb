# K-Means Clustering algorithm comparison

Two models are compared in this notebook: A manual Python K-Means Clustering algorithm, and a highly optimised Sklearn provided model.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import loadmat
from sklearn.cluster import KMeans
```

```python
bird_img = plt.imread('machine-learning-ex7/ex7/bird_small.png')
plt.imshow(bird_img)
plt.show()
```

```python
datafile = loadmat('machine-learning-ex7/ex7/bird_small.mat', struct_as_record=False)

X = datafile['A']
print("The shape of X is: {}".format(X.shape))

# flatten img data for input into our K Means model
X = X.reshape(X.shape[0]*X.shape[1], X.shape[2])
print("The new shape of X is: {}".format(X.shape))
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
plt.show()
```

```python
def K_means(K, X, iterations=10, print_progress=False):
    
    m = X.shape[0]
    
    # randomly initialise K centroids
    K_centroids = cluster_initialise(X, K, m)
    
    K_history = np.copy(K_centroids)
    costs = []
    
    for iteration in range(iterations):
        # assign each data point to the closest centroid
        idx = find_closest_centroids(X, K_centroids, m)
        
        # calculate optimisation objective
        cost = compute_cost(X, K_centroids, idx)
        costs.append(cost)
    
        # move centroids to the means of points allocated to each
        K_centroids = compute_means(X, K_centroids, idx)
        
        # store historic K centroid values over iterations
        K_history = np.append(K_history, K_centroids, axis=0)
        
        if print_progress:
            print("Iteration {0}: Centroid values are {1}".format(iteration, K_centroids))
            print("Cost is: {0}".format(cost))
        
    return K_centroids, idx, K_history, costs
    
        
def cluster_initialise(X, K, m):
    K_centroids = [np.random.choice(X[:, 0]) for k in range(K)]
    
    for i in range(1, X.shape[1]):
        K_ = [np.random.choice(X[:, i]) for k in range(K)]
        K_centroids = np.column_stack((K_centroids, K_))
    return K_centroids


def find_closest_centroids(X, K_centroids, m):
    indexes = []
    for i in range(m):
        k_dists = np.linalg.norm(X[i] - K_centroids, axis=1)**2
        closest_k = np.argmin(k_dists)
        indexes.append(closest_k)
    indexes = np.array(indexes)
    return indexes


def compute_means(X, K_centroids, idx):
    centroids = np.copy(K_centroids)
    for i in range(centroids.shape[0]):
        if X[idx==i].shape[0] > 0:
            centroids[i] = X[idx==i].sum(axis=0) / (X[idx==i].shape[0])
    return centroids


def compute_cost(X, K_centroids, idx):
    cost_sum = 0
    for i in range(X.shape[0]):
        cost_sum += np.linalg.norm(X[i] - K_centroids[idx[i]])**2
    cost = cost_sum / float(X.shape[0])
    return cost
```

### Visualise the clustering of this data in 3D and compare scratch model and sklearn model

#### Model built in Python from scratch

```python
centroids, idx, centroid_history, cost = K_means(16, X, 100, print_progress=False)
```

```python
X_reduced = np.copy(X)
# convert all data points so that their RGB value is equal to the assigned centroid
for i in range(centroids.shape[0]):
    X_reduced[idx==i] = centroids[i]

# ensure that we have 16 unique RGB colour combinations
np.unique(X_reduced, axis=0).shape
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=idx, alpha=0.3)
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], s=250, marker='*', c='r')
plt.title("K Means Clustering - built from scratch")
plt.show()
```

```python
# reshape flattened image into (128, 128, 3) and visualise compressed image
remade_img = X_reduced.reshape((128, 128, 3))

plt.imshow(remade_img)
plt.show()
```

#### Model built using Sklearn KMeans

```python
datafile = loadmat('machine-learning-ex7/ex7/bird_small.mat', struct_as_record=False)
X = datafile['A']
X = X.reshape(X.shape[0]*X.shape[1], X.shape[2])

kmeans_model = KMeans(n_clusters=16)
kmeans_model = kmeans_model.fit(X)

# the indexes that each data sample belong to are stored in kmeans.labels_
idx = kmeans_model.labels_

# the centroid locations are stored in kmeans.cluster_centers_
centroids = kmeans_model.cluster_centers_
```

```python
X_reduced = np.copy(X)
# convert all data points so that their RGB value is equal to the assigned centroid
for i in range(centroids.shape[0]):
    X_reduced[idx==i] = centroids[i]

# ensure that we have 16 unique RGB colour combinations
np.unique(X_reduced, axis=0).shape
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=idx, alpha=0.3)
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], s=250, marker='*', c='r')
plt.title("Sklearn K Means Clustering Model")
plt.show()
```

```python
# reshape flattened image into (128, 128, 3) and visualise compressed image
remade_img = X_reduced.reshape((128, 128, 3))

plt.imshow(remade_img)
plt.show()
```

#### Conclusions

The spread on our clusters is much better and more representative of the image in the sklearn model, when compared to the model built from scratch in Python. This is very apparent from inspecting the 3D plots of both, and visualising the spread of the clusters on the data.

Our manual model seems to be biased towards the larger values in the model, whereas the sklearn model has a much better spread with its clustering centroid locations. This is why the manual compressed image lacks many colours, whereas the sklearn model retains the colour qualities of the original image.

As a result, the sklearn model KMeans image compression is impressive; The image does not look bad, considering it is now $ \frac{1}{16} $ the size of the original image.
