### Anomaly Detection

So why do we use an anomaly detection algorithm, rather than supervised learning? The answer lies in how balanced our output classes (or labels) are with our data. 

If we only have a small number of anomalies, or if they're relatively difficult to classify compared to normal, non-anomalous data samples, then anomaly detection is a sound choice. 

For example, if we have 9800 good normal data samples, and only 200 anomalous samples, then we don't really have enough anomalous values for our model to learn effectively what anomalous samples should look like. In addition, our anomalies could vary greatly compared to one-another and therefore have many different types of anomalies. This makes it even harder for our algorithm to learn. In situations such as this, it is easier to use an anomaly detection model that knows what the data should look like, and identifies any outliers that do not fit the normal data trend.  

Given a training set $ \{x^{1}, x^{2}, x^{2},..., x^{m}\} $, where each sample, x, has n features: $ x \in \mathbb{R}^n $, the probability density estimation for a single example is given as:

$ p(x) = p(x_{1}, \mu_{1}, \sigma_{1}^{2})p(x_{1}, \mu_{1}, \sigma_{1}^{2})...p(x_{n}, \mu_{n}, \sigma_{n}^{2}) $

$ p(x) = \displaystyle\prod_{j=1}^{n} p(x_{j}, \mu_{j}, \sigma_{j}^{2}) $

Where in this case, the function p is the guassian density function, given by:

$ f(x) = \frac{1}{\sqrt{(2 \pi)} \sigma_{j}}exp\large(-\frac{(x_{j}-\mu_{j})^{2}}{2\sigma_{j}^{2}} \large) $

Where $ \mu $ is the mean, and $ \sigma $ the standard deviation. Also note that the variance is $ \sigma^{2} $

Notice that we're calculating the mean and variance for each feature column (with X of dimensions m x n, where m is samples of data and n is number of features (columns)), followed by computing p(x) for that sample of data. In practice, we can implement this much more efficiently in a vectorised form, where we calculate this using linear algebra very quickly in a library such as numpy.

#### Multivariate Gaussian

An alternative to the standard gaussian distribution model, as above, is the multivariate gaussian, which calculates p(x) in one go, rather than computing the individual feature p(x)'s and computing the product.

The probability density function for a multivariate gaussian (normal) is as follows:

$ f(x) = \frac{1}{\sqrt{(2 \pi)^{k} |\Sigma|}}exp\large(-\frac{1}{2}(x-\mu)^{T}\Sigma^{-1}(x-\mu) \large) $

Where $ \mu $ is the mean, $ \Sigma $ the covariance matrix, and k the number of feature dimensions of x. Note that $|\Sigma|$ is the determinant of the matrix $ \Sigma $, where $ \Sigma \in \mathbb{R}^{n \times n} $.

The covariance matrix can be thought of as the multidimensional analog of 1-d variance, and for multi-dimensional datasets its what we'll be using for a more efficient computation of the probability density function. 

The standard gaussian (univariate) and multivariate gaussians are the same if the covariance matrix has only diagonal components (the remainder of elements in the matrix being zero).

The benefit with using the multivariate model is that correlations between features will automatically be captured, rather than having to manually find good features in the standard model. The disadvantage is that it is much more computationally expensive due to the large matrix algebra involved, and the number of samples, m, must be greater than the number of features, n, as a minimum for the mathematics to work (otherwise $ \Sigma $ is non-invertable).

#### Mean and Standard Deviation Squared (Variance) Formulae:

$ \mu_{j} = \frac{1}{m} \displaystyle\sum_{i=1}^{m} x_{j}^{(i)} $

$ \sigma_{j}^{2} = \frac{1}{m} \displaystyle\sum_{i=1}^{m} x_{j}^{(i)} - \mu_{j} $

```python
def mean(X):
    return np.mean(X, axis=0)

def variance(X):
    return np.var(X, axis=0)
```

```python
mu = mean(X)
sigma2 = variance(X)

print("Mean: {0}, Variance: {1}".format(mu, sigma2))
```

```python
def multivariate_gaussian(X, mu, sigma2):
    """Computes the probability density function of the multivariate 
        gaussian distribution.
        If sigma2 is a vector > 1, it is treated as the covariance matrix."""
    
    k = len(mu)
    
    # if sigma2 is a vector of values, treat as covariance matrix
    if len(sigma2) > 1:
        sigma2 = np.diag(sigma2)
    
    # mean normalise the features of X
    X_ = np.copy(X)
    X_ -= mu.flatten()
    
    # multivariate gaussian computations
    term_1 = (((2*math.pi)**(-k/2))*np.linalg.det(sigma2))**-0.5
    term_2 = np.exp(-0.5*np.sum(np.multiply(X_, np.dot(X_, np.linalg.pinv(sigma2))), axis=1))
    
    return term_1 * term_2   


def gaussian_function(X, mu, sigma2):
    """Probability density gaussian function for single dimensional input data"""
    term_1 = ((2*math.pi*sigma2)**0.5)
    term_2 = np.exp(-0.5*(((X-mu)**2)/sigma2))
    return term_2 / term_1
```

#### Choose a value of epsilon for identifying anomalies, and highlight anomalous values on our plot

if $ p(x) >= \epsilon $, then y = 0 (normal)

#### Formally choosing the best value for epsilon - F1 score from cross-validation dataset

A better approach than randomly choosing epsilon is to select a threshold based on a metric or score of the cross-validation dataset, in this case we'll use the $ F_{1} $ score.

$ F_{1} $ is computed using the precision and recall values:

$$ F_{1} = \frac{2 \times precision \times recall}{precision + recall} $$

Where:

$$ Precision = \frac{true positives}{true positives + false positives} $$

$$ Recall = \frac{true positives}{true positives + false negatives} $$

```python
def select_threshold(y_val, p_val):
    """Select the best epsilon threshold value to use for outlier selection.
        Uses the ground truth (y_val) and validation set p_vals obtained from the
        multivariate gaussian probability density function. """
    
    # initialise 1000 epsilon values based on p_vals
    epsilons = np.arange(p_val.min(), 
                         p_val.max(), 
                         (p_val.max() - p_val.min()) / 1000)
    
    # create an array to store epsilon and F1 score pairs
    epsilon_F1_pairs = []
    
    for epsilon in epsilons:
        
        # classify anomaly/non-anomaly using epsilon
        y_preds = np.zeros(y_val.shape)
        y_preds[p_val < epsilon] = 1
        
        # calc true positives, false positives and false negatives
        true_positives = np.sum(np.logical_and(y_preds == 1, y_val == 1))
        false_positives = np.sum(np.logical_and(y_preds == 1, y_val == 0))
        false_negatives = np.sum(np.logical_and(y_preds == 0, y_val == 1))
        
        # calculate precision and recall - use small value to prevent div by 0
        precision = true_positives / (float(true_positives) + false_positives + 1e-12)
        recall = true_positives / (float(true_positives) + false_negatives + 1e-12)
        
        # calc F1 using precision and recall
        F1_score = (2*precision*recall) / (precision + recall + 1e-12)
        
        # append epsilon, F1 score to epsilon_F1_pairs
        epsilon_F1_pairs.append((epsilon, F1_score, y_preds))
    
    # sort by best F1 score
    epsilon_F1_pairs.sort(key=lambda x: x[1], reverse=True)
    
    return epsilon_F1_pairs[0][0], epsilon_F1_pairs[0][1], epsilon_F1_pairs[0][2]
```

Using our cross-validation dataset, X_val, we'll use the above function to calculate the F1 score for 1000 different epsilon values, and select the best scoring epsilon.

```python
# calculate p(x) for val data, using mu/sigma2 found from training set, X
validation_p_vals = multivariate_gaussian(X_val, mu, sigma2)

# get top epsilon, F1 score, y predictions
top_epsilon, F1_score, y_val = select_threshold(y_val, validation_p_vals)
```

```python
print("The best epsilon value found was: {}.".format(top_epsilon))
```

#### Plot a simple confusion matrix for our top epsilon value

```python
import itertools
from sklearn.metrics import confusion_matrix

# classify anomaly/non-anomaly using epsilon
y_preds = np.zeros(y_val.shape)
y_preds[validation_p_vals < top_epsilon] = 1

conf_matrix = confusion_matrix(y_val, y_preds)
conf_matrix

def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """ This function plots a simple confusion matrix. """

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
plt.figure(figsize=(6, 5))
plot_confusion_matrix(conf_matrix, classes=['normal', 'anomalous'],
                      title='Confusion matrix - anomaly detector epsilon of {0:.5f}'.format(top_epsilon))
plt.show()
```
